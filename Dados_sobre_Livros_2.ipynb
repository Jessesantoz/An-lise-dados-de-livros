# Análise de dados gerais sobre livros


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dados/books.csv', on_bad_lines='skip')

df.dropna()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

# Publicações por Língua

```python
lingua = df.filter(items=['language_code', 'bookID']).groupby(['language_code']).count().sort_values(by='bookID')
```

```python
lingua
```

```python
lingua.plot(kind='bar', color='red')
plt.title('DISTRIBUIÇÃO DOS LIVROS POR LÍNGUA:')
plt.ylabel('Quantidade')
plt.xlabel('Língua')
plt.show()
```

# Principais Autores

```python
autoresLivros = df.filter(items=['authors', 'bookID']).groupby(['authors']).count().sort_values(by='bookID', ascending=True)
```

```python
autoresLivros.tail(3)
```

```python
topAutoresLivros = autoresLivros.tail(10)
```

```python
topAutoresLivros.plot(kind='barh', color='black')
plt.title('TOP 10 AUTORES POR LIVRO')
plt.show()
```

```python
autoresAvaliacao = df.filter(items=['authors', 'ratings_count']).groupby(['authors']).count().sort_values(by='ratings_count', ascending=True)
```

```python
autoresAvaliacao.tail(3)
```

```python
topautoresAva = autoresAvaliacao.tail(10)
```

```python
topautoresAva.plot(kind='barh', color='green')
plt.title('TOP 10 AUTORES POR AVALIAÇÃO')
plt.show()
```

```python
autoresVisu = df.filter(items=['authors', 'text_reviews_count']).groupby(['authors']).count().sort_values(by='text_reviews_count', ascending=True)
```

```python
autoresVisu.tail(3)
```

```python
topautoresVisu = autoresVisu.tail(10)
```

```python
topautoresVisu.plot(kind='barh', color='red')
plt.title('TOP 10 AUTORES POR VISUALIZAÇÃO')
plt.show()
```

```python
df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
```
