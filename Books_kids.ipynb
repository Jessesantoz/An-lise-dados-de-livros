# Avaliação

```python
condicao = df['Rating_out_of_5'] > 4
```

```python
df[condicao]
```

```python
maximo = df['Rating_out_of_5'] == df['Rating_out_of_5'].max()
df_maior = df[maximo]
df_maior['Name'].iloc[0]
```

```python
entre = (df['Rating_out_of_5'] >= 3) & (df['Rating_out_of_5'] <= 4.5)
df[entre]
```

```python
df_avaliacao = df.groupby(['Author']).agg({'Rating_out_of_5': "sum",
       'No_of_Ratings': 'count'}).reset_index()

df_avaliacao['valor_autor'] = df['Rating_out_of_5'] / df['No_of_Ratings']

df_avaliacao.sort_values(by='valor_autor', ascending=False)
```
