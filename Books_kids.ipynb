# Principais obras

```python
df['Name'].value_counts()
```

```python
obras = df.filter(items=['Name', 'No_of_Ratings']).sort_values(ascending=False, by='No_of_Ratings').head(10)
obras
```

```python
top = df.filter(items=['Name', 'Description', 'Author', 'Age', 'Rating_out_of_5','Price', 'Price_Befor',
                       'Cover_Type','Publication_Date', 'Product_Details', 'Best_Seller'])\
                       .where(df.Name.isin(['Amazing Spider-Man (1999 - 2013)', 'Iron Man (1968 - 1996)'
                       , 'Captain America (1968 - 1996)', 'Iron Man (1968 - 1996)', 'Captain America (1968 - 1996)',
                       'Iron Man (1968 - 1996)', 'Captain America (1968 - 1996)', 'Iron Man (1968 - 1996)',
                        'Captain America (1968 - 1996)']))
```

```python
top = df.filter(items=['Name', 'Description', 'Author', 'Age', 'No_of_Ratings', 'Rating_out_of_5','Price', 'Price_Befor',
                       'Cover_Type','Publication_Date', 'Best_Seller']).where(df['No_of_Ratings'] >= 971.0)\
                       .sort_values(ascending=False, by='No_of_Ratings').head(10)

top
```

# Datas

```python
data = df['Publication_Date'].value_counts().sort_index(ascending=False)
```

```python
data.plot(color='green')
plt.legend(bbox_to_anchor=(1.0, 1), loc=0)
plt.rcParams['figure.figsize'] = (15, 5)
plt.title ('QUANTIDADE DE PRODUÇÃO POR ANO')
plt.show()
```
